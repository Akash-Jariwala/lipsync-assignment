## Display Resultant Video

```python
# diaplay output video
from IPython.display import display
from google.colab import output
import ipywidgets as widgets
import base64

def play_video(video_path):
    video_file = open(video_path, "rb").read()
    video_encoded = base64.b64encode(video_file).decode('utf-8')
    video_tag = '<video controls alt="video" src="data:video/mp4;base64,{0}">'.format(video_encoded)
    output.clear()
    display(widgets.HTML(video_tag))

# ==== Provide the path to your MP4 file ====

# -- if you are running notebook locally then... --
# video_path = 'results/result_voice.mp4'

# if you are using colab then...
video_path = '/content/Wav2Lip/results/result_voice.mp4'

play_video(video_path)
```
